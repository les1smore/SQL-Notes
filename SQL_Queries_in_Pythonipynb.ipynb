```python
# Import the library and moduel
from pandasql import sqldf
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/17rsuraj/data-curious/master/TowardsDataScience/Dummy_Sales_Data_v1.csv', sep =',')
df.head()
```

```python
# Rename the columns for easy accessibility
df.rename(columns ={'Shipping_Cost(USD)': 'ShippingCost_USD',
                    'UnitPrice(USD)':'UnitPrice_USD',
                    'Delivery_Time(Days)':'Delivery_Time_Days'}, inplace = True)
df.info()
```

# Data Extraction

```python
# Make a table named df_orders
query = 'SELECT OrderID, Quantity, Sales_Manager, Status, Shipping_Address, ShippingCost_USD FROM df'
df_orders = sqldf(query)
df_orders.head()
```

```python
# Make a table named df_products
query = 'SELECT OrderID, Quantity, Product_Code, Product_Category, UnitPrice_USD FROM df'
df_products = sqldf(query)
df_products.head()
```

```python
# Use multi-line string
query = "SELECT OrderID,\
        Quantity, \
        Product_Code, \
        Product_Category, \
        UnitPrice_USD \
        FROM df"
df_products = sqldf(query)
df_products.head()
```

# Select Data based on Condition

```python
# Select all orders from df_orders where the shipping address is Kenya
query = 'SELECT *\
        FROM df_orders \
        WHERE Shipping_Address = "Kenya"'
df_kenya = sqldf(query)
df_kenya.head()
```

```python
# Select orders shipped to Kenya and had quantity less than 40 and have status as shipped or delivered
query = "SELECT * \
        FROM df_orders \
        WHERE Shipping_Address = 'Kenya' \
        AND Quantity < 40 \
        AND Status IN ('Shipped', 'Delivered')"
```

```python
df_kenya = sqldf(query)
df_kenya.head()
```

# Groupby

```python
# How many orders are delivered to each shipping address
query = "SELECT Shipping_Address, \
        COUNT(OrderID) AS Orders \
        FROM df_orders \
        GROUP BY Shipping_Address"
df_group = sqldf(query)
df_group.head()
```

`sqldf` gives out pandas DataFrame as output whereas pandas groupby returns DataFrame Object as output.

# Orderby

```python
query = 'SELECT Shipping_Address, \
        COUNT(OrderID) AS Orders \
        FROM df_orders \
        GROUP BY Shipping_Address \
        ORDER BY Orders'
df_group = sqldf(query)
df_group
```

# Combine the Datasets

```python
# Get all the orders, its shipping address and product category together
query = "SELECT T1.OrderID, \
        T1.Shipping_Address, \
        T2.Product_Category \
        FROM df_orders T1\
        INNER JOIN df_products T2\
        ON T1.OrderID = T2.OrderID"
df_combined = sqldf(query)
df_combined.head()
```
